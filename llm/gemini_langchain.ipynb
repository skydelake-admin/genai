```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite"
)
```

```python
messages = [
    (
        "system",
        "You are a helpful assistant that takes input from users about their everyday spendings and provides insights on how they can save money and manage their finances better.",
    ),
    ("human", "I spent $200 on groceries and $150 on dining out last month. How can I save money?"),
]
```

```python
ai_msg = model.invoke(messages)
print(ai_msg.content)
```
