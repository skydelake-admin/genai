```python
from langchain_openrouter import ChatOpenRouter
```

```python
import os
from dotenv import load_dotenv

# Load local .env file (do not commit .env with secrets to public repos)
load_dotenv()

# Ensure you set your key first: create a .env with OPENROUTER_API_KEY or set env
api_key = os.getenv("OPENROUTER_API_KEY")
if not api_key:
	print("Set OPENROUTER_API_KEY in the environment to run this cell.")
else:
	model_name = os.getenv("OPENROUTER_MODEL", "openai/gpt-4o-mini")
	model = ChatOpenRouter(model=model_name)
	prompt = "When was the first moon landing?"
	resp = model.invoke(prompt)
	print(getattr(resp, 'content', getattr(resp, 'text', resp)))
	
```
