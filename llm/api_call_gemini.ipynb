# Gemini API Connection

This notebook shows how to call the Gemini API using your API key stored in an environment variable.

```python
import os
from pathlib import Path
from dotenv import load_dotenv
import requests

# Load .env from the repository root so GEMINI_API_KEY is available in the notebook
env_path = Path.cwd() / '.env'
if not env_path.exists():
    env_path = Path.cwd().parent / '.env'
load_dotenv(env_path)

api_key = os.getenv('GEMINI_API_KEY')
# print('Loaded .env from:', env_path)
# print('GEMINI_API_KEY visible in kernel:', bool(api_key))
if not api_key:
    raise ValueError('Missing GEMINI_API_KEY. Make sure .env contains GEMINI_API_KEY.')

url = 'https://generativelanguage.googleapis.com/v1/models/gemini-2.5-flash:generateContent'
headers = {'Content-Type': 'application/json'}
payload = {
    'contents': [
        {
            'role': 'user',
            'parts': [
                {'text': 'what is temperature in gen ai LLM models?'}
            ]
        }
    ],
    'generationConfig': {
        'temperature': 0.2,
        'maxOutputTokens': 12000,
    }
}

response = requests.post(url, headers=headers, params={'key': api_key}, json=payload)
response.raise_for_status()
data = response.json()

# print('Raw response keys:', list(data.keys()) if isinstance(data, dict) else type(data))

output_text = None
if isinstance(data, dict):
    candidates = data.get('candidates') or []
    if candidates:
        content = candidates[0].get('content') or {}
        parts = content.get('parts') or []
        output_text = ''.join([p.get('text', '') for p in parts if isinstance(p, dict) and p.get('text')])

if output_text:
    print('\LLM response:')
    print(output_text)
else:
    print('Could not find the generated text in the response.')
```
